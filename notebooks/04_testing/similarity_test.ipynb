```python
# Generate embeddings in Python
import ollama

from numpy import dot
from numpy.linalg import norm

paragraph1 = "FABRIC site mentions"
paragraph2 = "I cannot help you with that"
```

```python
#model='nomic-embed-text'
# model = 'all-minilm:latest'
models = ['nomic-embed-text', 'all-minilm:latest']
```

```python
for model in models:
    # Get embeddings
    emb1 = ollama.embeddings(model=model, prompt=paragraph1)
    emb2 = ollama.embeddings(model=model, prompt=paragraph2)

    # Calculate cosine similarity

    similarity = dot(emb1['embedding'], emb2['embedding']) / (norm(emb1['embedding']) * norm(emb2['embedding']))
    print(f"Similarity: {similarity}")
```
